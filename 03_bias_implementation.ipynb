```python
import tensorflow as tf

w = tf.Variable([2.0])
b = tf.Variable([1.0])

def linear_model(x):
    return w * x + b
```

```python
print("predicted output when input is 0")
result1 = linear_model(tf.constant([0.0]))
print(result1.numpy())
```

```python
print("\npredicted output when input is 0")
result2 = linear_model(tf.constant([0.0, 0.0, 0.0]))
print(result2.numpy())
```

```python
print("\nPredictions with inputs [1,2,3,4]")
result3 = linear_model(tf.constant([1.0, 2.0, 3.0, 4.0]))
print(result3.numpy())
```
