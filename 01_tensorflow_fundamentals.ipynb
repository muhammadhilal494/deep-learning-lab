```python
import tensorflow as tf

a = tf.constant(0.3, dtype=tf.float32)  # or any initial value
b = tf.constant(0.5, dtype=tf.float32)

c = a + b
print(c)
```

```python
import tensorflow as tf

def compute(a, b):
    return a + b

c = compute(1, 4)
print(c)
```

```python
import tensorflow as tf

a = tf.constant([1, 2, 3], dtype=tf.float32)
b = tf.constant([4, 2, 1], dtype=tf.float32)

c = a + b
print(c)
```

```python
import tensorflow as tf

x = tf.Variable([1.0])
w = tf.Variable([3.0])

c = x * w

print(c)
```
