```python
import tensorflow as tf
import os

# TensorFlow 2.x version
w = tf.Variable([0.2])   # initial weight
b = tf.Variable([-0.2])  # initial bias

# Instead of placeholder, define as function parameter
def linear_model(x):
    return w * x + b
```

```python
# ✅ TensorFlow 2.x - Direct execution:
print("Weight w:", w.numpy())
print("Bias b:", b.numpy())
```

```python
print("predicted output is")
result = linear_model(tf.constant([1.0, 2.0, 3.0, 4.0]))
print(result.numpy())
```
