```python
import tensorflow as tf
import os

# TensorFlow 2.x version
w = tf.Variable([0.2])   # initial weight
b = tf.Variable([-0.2])  # initial bias

# Instead of placeholder, define as function parameter
def linear_model(x):
    return w * x + b
```

```python
# ✅ TensorFlow 2.x - Direct execution:
print("Weight w:", w.numpy())
print("Bias b:", b.numpy())
```

```python
print("predicted output is")
result = linear_model(tf.constant([1.0, 2.0, 3.0, 4.0]))
print(result.numpy())
```

```python
def calculate_loss(x, y):
    predictions = linear_model(x)
    squared_deltas = tf.square(predictions - y)
    loss = tf.reduce_sum(squared_deltas)  # sum all squared errors
    return loss
```

```python
x = tf.constant([1.0, 2.0, 3.0, 4.0])
y = tf.constant([0.0, -1.0, -2.0, -3.0])\
```

```python
loss_value = calculate_loss(x, y)
print("loss or Error is")
print(loss_value.numpy())
```
