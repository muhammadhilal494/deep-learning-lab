```python
import tensorflow as tf
import os

w = tf.Variable([0.2])
b = tf.Variable([-0.2])

def linear_model(x):
    return w * x + b
```

```python
print("Weight w:", w.numpy())
print("Bias b:", b.numpy())
```

```python
print("predicted output is")
result = linear_model(tf.constant([1.0, 2.0, 3.0, 4.0]))
print(result.numpy())
```

```python
def calculate_loss(x, y):
    predictions = linear_model(x)
    squared_deltas = tf.square(predictions - y)
    loss = tf.reduce_sum(squared_deltas)
    return loss
```

```python
x = tf.constant([1.0, 2.0, 3.0, 4.0])
y = tf.constant([0.0, -1.0, -2.0, -3.0])
```

```python
loss_value = calculate_loss(x, y)
print("loss or Error is")
print(loss_value.numpy())
```

```python
optimizer = tf.optimizers.SGD(learning_rate=0.01)

x_train = tf.constant([1.0, 2.0, 3.0, 4.0])
y_train = tf.constant([0.0, -1.0, -2.0, -3.0])

print("error is 20.16. So lets update the weights and check error again")
print("=" * 60)

for epoch in range(1000):
    with tf.GradientTape() as tape:
        loss = calculate_loss(x_train, y_train)

    gradients = tape.gradient(loss, [w, b])
    optimizer.apply_gradients(zip(gradients, [w, b]))

print("Training Complete!")
print("=" * 60)
```

```python
print("\nwe will get values for weight and bias")
print("wight: ", w.numpy())
print("Bias: ", b.numpy())
```

```python
print("\nPredictions after training:")
predictions = linear_model(tf.constant([1.0, 2.0, 3.0, 4.0]))
print(predictions.numpy())
```

```python
print("\nthese values are near the desired output")

final_loss = calculate_loss(x_train, y_train)
print(f"\nFinal RMSE: {final_loss.numpy()}")
```

```python
print(f"\nRMSE after gradient descent optimizer is {final_loss.numpy():.10e}")
print("Earlier we were getting 20.16")
```
