```python
X = df[df.columns[0:60]].values
print("Features shape:", X.shape)
print(X)
```

```python
y = df[df.columns[60]]
print("\nLabels shape:", y.shape)
print("Unique labels:", y.unique())
print(y)
```

```python
from sklearn.preprocessing import LabelEncoder

y_labelencoder = LabelEncoder()
y = y_labelencoder.fit_transform(y)
print(y)
```

```python
y = y.reshape(-1, 1)
print("Reshaped y:", y.shape)

onehotencoder = OneHotEncoder(sparse_output=False)
Y = onehotencoder.fit_transform(y)

print("One-Hot Encoded Y:")
print(Y)
```

```python
train_x, test_x, train_y, test_y = train_test_split(
    X, Y, test_size=0.20, random_state=42
)

train_x = tf.cast(train_x, tf.float32)
test_x = tf.cast(test_x, tf.float32)
train_y = tf.cast(train_y, tf.float32)
test_y = tf.cast(test_y, tf.float32)

# Print shapes
print("Training X shape:", train_x.shape)
print("Training Y shape:", train_y.shape)
print("Testing X shape:", test_x.shape)
print("Testing Y shape:", test_y.shape)
```

```python
# Hyperparameters
learning_rate = 0.01
training_epochs = 100
n_class = 2
```

```python
# Initialize histories
cost_history = []
mse_history = []
```

```python
# Model variables
W = tf.Variable(tf.zeros([60, n_class], dtype=tf.float32))
b = tf.Variable(tf.zeros([n_class], dtype=tf.float32))
```

```python
# Model function
def perceptron_model(x):
    logits = tf.matmul(x, W) + b
    pred = tf.nn.softmax(logits)
    return pred, logits
```

```python
# Loss function
def loss_fn(y_true, logits):
    return tf.reduce_mean(
        tf.nn.softmax_cross_entropy_with_logits(y_true, logits)
    )
```

```python
# Optimizer
optimizer = tf.optimizers.SGD(learning_rate=learning_rate)
```

```python
# ===== Training Loop =====
print("Training the Sonar Perceptron Model...")
print("=" * 60)

for epoch in range(training_epochs):
    # Train on training data
    with tf.GradientTape() as tape:
        _, logits_train = perceptron_model(train_x)
        cost = loss_fn(train_y, logits_train)

    # Compute gradients and update weights
    gradients = tape.gradient(cost, [W, b])
    optimizer.apply_gradients(zip(gradients, [W, b]))

    # Store cost
    cost_history.append(cost.numpy())

    # Evaluate on test data
    pred_y, _ = perceptron_model(test_x)

    # Calculate MSE
    mse = tf.reduce_mean(tf.square(pred_y - test_y))
    mse_history.append(mse.numpy())

    # Print progress
    print(f"epoch: {epoch:3d} - cost: {cost.numpy():.6f} - MSE: {mse.numpy():.6f}")

print("=" * 60)
print("Training Complete! ✅")
```

```python
def calculate_accuracy(pred, y_true):
    correct_prediction = tf.equal(tf.argmax(pred, 1), tf.argmax(y_true, 1))
    accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
    return accuracy.numpy()
```

```python
# Get predictions on test data
pred_test, _ = perceptron_model(test_x)

# Calculate accuracy
accuracy = calculate_accuracy(pred_test, test_y)
print("Accuracy:", accuracy)
```
